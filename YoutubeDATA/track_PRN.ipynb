```python
import pandas as pd
from googleapiclient.discovery import build
from datetime import datetime
```

```python
# Global popluar right now 재생목록 비디오 아이디 받아오기
f = open("key.txt",'r')
DEVELOPER_KEY= f.readline()
f.close()

# Create a YouTube API client
youtube = build('youtube', 'v3', developerKey=DEVELOPER_KEY)
playlistId = 'PLrEnWoR732-BHrPp_Pm8_VleD68f9s14-'
today = datetime.today()
currentDay = today.strftime("%Y%m%d")

playlist_items = []
next_page_token = None

while True:
    # Retrieve the next page of playlist items
    playlist_request = youtube.playlistItems().list(
        part='contentDetails',
        playlistId=playlistId,
        maxResults=50,
        pageToken=next_page_token
    )
    playlist_response = playlist_request.execute()

    # Add the video IDs to the list
    for item in playlist_response['items']:
        playlist_items.append(item['contentDetails']['videoId'])

    # Check if there are more pages to retrieve
    next_page_token = playlist_response.get('nextPageToken')
    if not next_page_token:
        break
category_id=[]
views=[]
likes=[]
favorites = []
comments=[]
title=[]
date=[]
tag=[]
for videoId in playlist_items:
    request=youtube.videos().list(
    part='snippet,statistics',
    id=videoId,
    maxResults=100)
    
    response=request.execute()
    
    if response['items']==[]:
        title.append('-')
        category_id.append('-')
        views.append('-')
        likes.append('-')
        favorites.append('-')
        comments.append('-')
        date.append('-')
        tag.append('-')
        
    else :
        title.append(response['items'][0]['snippet']['title'])
        category_id.append(response['items'][0]['snippet']['categoryId'])
        views.append(response['items'][0]['statistics']['viewCount'])
        if 'likeCount' in response['items'][0]['statistics']:
            likes.append(response['items'][0]['statistics']['likeCount'])
        else:
            likes.append('Nan')
        favorites.append(response['items'][0]['statistics']['favoriteCount'])
        if 'comments' in response['items'][0]['statistics']:
            comments.append(response['items'][0]['statistics']['commentCount'])
        else:
            comments.append('Nan')
        date.append(response['items'][0]['snippet']['publishedAt'])
        if 'tags' in response['items'][0]['snippet']:
            tag.append(response['items'][0]['snippet']['tags'])
        else:
            tag.append('Nan')
    
df=pd.DataFrame([title,category_id,views,likes,favorites,comments,date,tag]).T
df.columns=['title','category_id','views','likes','favorites','comments','date','tags']
today = datetime.today()
currentDay = today.strftime("%Y%m%d")
df.to_csv(f'./DATA/PopularRightNow/PRN_global_{currentDay}.csv', sep=',', na_rep='NaN')
```
