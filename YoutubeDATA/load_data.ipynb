```python
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from oauth2client.tools import argparser
import pandas as pd
```

```python
f = open("key.txt",'r')
DEVELOPER_KEY= f.readline()
f.close()

YOUTUBE_API_SERVICE_NAME='youtube'
YOUTUBE_API_VERSION='v3'

youtube=build(YOUTUBE_API_SERVICE_NAME,YOUTUBE_API_VERSION,developerKey=DEVELOPER_KEY)
```

```python
#í”Œë ˆì´ë¦¬ìŠ¤íŠ¸ì—ì„œ ê° ë™ì˜ìƒì˜ idë¥¼ ë°›ì•„ ë¦¬ìŠ¤íŠ¸ë¡œ ì œì‘

videoIds = []
playListIds = ['PL7eXornP6T5RD2Nc5fyPxW_t6PVE2OEQw']

for i in playListIds:
    playlist_videos=youtube.playlistItems().list(
    playlistId=i,
    part='snippet',
    maxResults = 50)
    playlistitems_list_response=playlist_videos.execute()
    nextPageToken = playlistitems_list_response.get('nextPageToken')
    
    while ('nextPageToken' in playlistitems_list_response):
        nextPage = youtube.playlistItems().list(
        playlistId=i,
        part='snippet',
        maxResults = 50,
        pageToken = nextPageToken).execute() 
        playlistitems_list_response['items'] = playlistitems_list_response['items'] + nextPage['items']
    
        if 'nextPageToken'in nextPage:
            nextPageToken = nextPage['nextPageToken']
        else:
            playlistitems_list_response.pop('nextPageToken',None)

    for v in playlistitems_list_response['items']:
        videoIds.append(v['snippet']['resourceId']['videoId'])
        
videoIds
```

```python
#ì˜ìƒ ê´€ë ¨ ìˆ˜ì¹˜
category_id=[]
views=[]
likes=[]
favorites = []
comments=[]
title=[]
date=[]

for u in range(len(videoIds)):
    request=youtube.videos().list(
    part='snippet,contentDetails,statistics',
    id=videoIds[u],
    maxResults=100)
    
    response=request.execute()
    
    if response['items']==[]:
        title.append('-')
        category_id.append('-')
        views.append('-')
        likes.append('-')
        favorites.append('-')
        comments.append('-')
        date.append('-')
        
    else :
        title.append(response['items'][0]['snippet']['title'])
        category_id.append(response['items'][0]['snippet']['categoryId'])
        views.append(response['items'][0]['statistics']['viewCount'])
        likes.append(response['items'][0]['statistics']['likeCount'])
        favorites.append(response['items'][0]['statistics']['favoriteCount'])
        comments.append(response['items'][0]['statistics']['commentCount'])
        date.append(response['items'][0]['snippet']['publishedAt'])
```

```python
df=pd.DataFrame([title,category_id,views,likes,comments,date]).T
df.columns=['title','category_id','views','likes','comments','date']
df
```
